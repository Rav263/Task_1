```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = plt.imread('1.jpg')
```

```python
plt.imshow(img)
```

```python
w, h, c = img.shape
```

```python
print(w, h, c)
```

```python
 imgscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(imgscale, 'gray')
```

```python
w, h = imgscale.shape

w = 64 / w
h = 64 / h

imgscale = cv2.resize(imgscale, None, fx=h, fy=w, interpolation=cv2.INTER_NEAREST)
```

```python
plt.imshow(imgscale, 'gray')
```

```python
imgscale = cv2.resize(imgscale, None, fx=0.125, fy=0.125, interpolation=cv2.INTER_NEAREST)

plt.imshow(imgscale, 'gray')
```

```python
w, h = imgscale.shape
```
