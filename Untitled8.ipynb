```python
import requests
```

```python
url = requests.get("https://www.cian.ru/cat.php?deal_type=sale&engine_version=2&offer_type=flat&p=1&region=1")
```

```python
def get_flat_urls(url):
    text = url.text
    some = 0

    urls = []
    
    while some != -1:
        some = text.find("https://www.cian.ru/sale/flat/", some + 1)
        #print(some)
        now = text[some:some+40:1]
        #print(now)
        urls.append(now)
        
    return urls
```

```python
flats = []
for x in range(1, 60):
    url = "https://www.cian.ru/cat.php?deal_type=sale&engine_version=2&offer_type=flat&p="+str(x)+"&region=1"
    page = requests.get(url)
    print(url)
    
    flats.extend(get_flat_urls(page))
```

```python
file = open("flats.txt", "w")
```

```python
for x in flats:
    if len(x) == 0:
        continue
    file.write(x)
    file.write("\n")
```
