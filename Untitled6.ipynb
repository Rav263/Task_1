```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = plt.imread('some1.png')
```

```python
imggr = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
w, h = imggr.shape

w = 32 / w
h = 32 / h

imgscale = cv2.resize(imggr, None, fx=h, fy=w, interpolation=cv2.INTER_NEAREST)
plt.imshow(imgscale)
```

```python
mid = 0.0
for x in imgscale:
    print(x)
    for y in x:
        mid += y
mid /= 1024.0
```

```python
print(mid)
```

```python
imgbin = [[1 if imgscale[y][x] > mid else 0 for x in range(0, 32)] for y in range(0, 32)]
```

```python
plt.imshow(imgbin)
```

```python
img2 = plt.imread('1.jpg')
```

```python
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

plt.imshow(img2)
```

```python
imgblur = cv2.medianBlur(img2, 5)
plt.imshow(imgblur)
```

```python
w, h = imgblur.shape

w = 32 / w
h = 32 / h

img2res = cv2.resize(imgblur, None, fx=h, fy=w, interpolation=cv2.INTER_NEAREST)
```

```python
plt.imshow(img2res)
```

```python
dct2 = cv2.dct(np.float32(img2res))
```

```python
plt.imshow(np.uint8(dct2))
```

```python
mid = 0
for x in dct2:
    print(x)
    for y in x:
        mid += y

mid /= 1024
        
        
dct2 = [[1 if y > mid else 0 for y in x] for x in dct2]
print(mid)
```

```python
plt.imshow(dct2)
```

```python
hash = 0
for x in range(0, 8):
    for y in range(0, 8):
        hash <<= 1
        hash += dct2[x][y]
```

```python
print(hex(hash))
```
