```python
import requests
import time
```

```python
def get_flat_urls(url):
    text = url.text
    some = 0

    urls = []
    
    while some != -1:
        some = text.find("https://www.cian.ru/sale/flat/", some + 1)
        #print(some)
        now = text[some:some+40:1]
        #print(now)
        urls.append(now)
        
    return urls
```

```python
def get_images_urls(url):
    text = url.text
    some = text.find("window._cianConfig")
    part = text[some:some+23000:1]
    parts = part.split(":")
    
    images = []
    counter = 10
    last = ""
    for x in parts:
        if x.find(".jpg")==-1:
            continue
            
        now = x.split("\"")[0]
        now = "https:" + now
        now = now.replace("\\u002F", '\u002F')
        
        now_splited = now.split("-")
        now_indefer = now_splited[len(now_splited) - 2]
        if(now_indefer != last):
            images.append(now)
            #print(counter, now)
            last=now_indefer
    return images
```

```python
flats = []
for x in range(1, 60):
    url = "https://www.cian.ru/cat.php?deal_type=sale&engine_version=2&offer_type=flat&p="+str(x)+"&region=1"
    page = requests.get(url)
    now = get_flat_urls(page)
    flats.extend(now)
    print(x, url, len(now))
    time.sleep(1)
```

```python
images = []
for x in range(0, len(flats)):
    if len(flats[x]) == 0:
        continue
    page = requests.get(flats[x])
    now = get_images_urls(page)
    print(x, flats[x], len(now))
    images.extend(now)
```
