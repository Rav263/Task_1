```python
import cv2
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('index.jpg',0)
```

```python
orb = cv2.ORB_create()
```

```python
kp = orb.detect(img,None)
```

```python
kp, des = orb.compute(img, kp)
```

```python
img2 = cv2.drawKeypoints(img,kp,None,color=(0,255,0), flags=0)
plt.imshow(img2),plt.show()
```
