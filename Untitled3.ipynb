```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def hash_img (img):
    imggr = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    w, h = imggr.shape
    w = 64 / w
    h = 64 / h

    imgscale = cv2.resize(imggr, None, fx=h,  fy=w, interpolation=cv2.INTER_NEAREST)
    imgscale = cv2.resize(imgscale, None, fx=0.125,  fy=0.125, interpolation=cv2.INTER_NEAREST)
    
    imgscfl = np.float32(imgscale) / 255.0
    dct = cv2.dct(imgscfl)
    
    dctres = [[dct[x][y] for y in range(0, 8)] for x in range(0, 8)]
    
    mid = 0
    for x in range(0, 8):
        for y in range(0, 8):
            if x == 0 and y == 0: 
                continue
            mid += dctres[x][y]
    
    mid /= 63
    
    imgobr = cv2.idct(dct)
    
    plt.imshow(imgobr)
    
    dctbin = [[0 if dctres[x][y] < mid else 1 for y in range(0, 8)] for x in range(0, 8)]
    
    dctbinb = [[dctbin[x][y] if x < 8 and y < 8 else 0 for y in range(0, 32)] for x in range(0, 32)]
    dctbind = [[255 if dctbinb[x][y] == 1 else -255 for y in range(0, 32)] for x in range(0, 32)]
    
    imgdct = cv2.idct(np.float32(dctbinb))
    
    
    #plt.imshow(imgdct, 'gray')
    
    #for x in dctbinb:
       # print(x)
    
    hash = 0
    for x in dctbin:
        #print(x)
        for y in x:
            hash += y
            hash <<= 1
    return hash
```

```python
def midhash_img(img):
    imgscale = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    w, h = imgscale.shape

    w = 64 / w
    h = 64 / h

    imgscale = cv2.resize(imgscale, None, fx=h, fy=w, interpolation=cv2.INTER_NEAREST)
    imgscale = cv2.resize(imgscale, None, fx=0.125, fy=0.125, interpolation=cv2.INTER_NEAREST)
    
    #plt.imshow(imgscale)
    
    mid = 0
    for x in range(0, 8):
        for y in range(0, 8):
            mid += imgscale[x][y]
    
    mid /= 64
    
    imgbin = [[0 if imgscale[x][y] < mid else 1 for y in range(0, 8)] for x in range(0, 8)]
    
    #plt.imshow(imgbin)
    
    hash = 0;
    for x in imgbin:
        #print(x)
        for y in x:
            hash += y
            hash <<= 1
    return hash
```
