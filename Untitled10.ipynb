```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from tqdm import tqdm_notebook
```

```python
def hash_img(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.medianBlur(img, 5)
    
    w, h = img.shape

    w = 64 / w
    h = 64 / h

    img = cv2.resize(img, None, fx=h, fy=w, interpolation=cv2.INTER_LINEAR)
    
    dct = cv2.dct(np.float32(img))
    
    mid = 0
    for x in range(0, 8):
        for y in range(0, 8):
            mid += dct[x][y]

    mid /= 64
        
        
    dct = [[1 if y > mid else 0 for y in x] for x in dct]
    
    hash = 0
    for x in range(0, 8):
        for y in range(0, 8):
            hash <<= 1
            hash += dct[x][y]
            
    return hash
```

```python
hashs=[0]
```

```python
for x in tqdm_notebook(range(0, 15000)):
    if x == 5185 or x == 5188:
        hashs.append(0)
        continue
    now_img = plt.imread("./images/" + str(x) + '.jpg')
    hashs.append(hash_img(now_img))
```

```python
print(len(hashs))
```

```python
for x in tqdm_notebook(range(1, 1000)):
    for y in tqdm_notebook(range(1, 1000)):
        now = str(bin(hashs[x]^hashs[y])).count('1')
        hemmings.append(now)
```
